```python
df['Calories'] = df['Calories'].abs()
print(df) #remove the negative values from calories
```

```python
x = df['Calories'].mean() #take mean value of calories column in x
print('mean : ',x)
print('----------------------------------------------------------------------')
df['Calories'].fillna(x,inplace = True) #fill every null value with value of x
print(df)
```

```python
df.dropna(subset = ['Date'], inplace=True) #row with null value get's deleted from dataset
print(df)
```

```python
df['Date']=pd.to_datetime(df['Date']) #convert string data types Datecolumn to date type
print(df)
```

```python
df['Calories'] = df['Calories'].astype(int) #remove decimal from calories column
print(df)
```

```python
df.loc[7, 'Duration'] = 45 #change 7th index's Duratoin column's value to 45
print(df)
```

```python
df.duplicated().sum() #contains one duplicate entry
```

```python
df.drop_duplicates(inplace = True) #drop the row with duplicate
print(df) 
df.duplicated().sum()
```

```python
for x in df.index:
    if df.loc[x,'Duration']>45 :
        df.loc[x,'Duration']=50
print(df)
```
